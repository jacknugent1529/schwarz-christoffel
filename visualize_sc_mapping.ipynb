```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from tqdm import tqdm
tmp = np.loadtxt("build/out.txt", delimiter=",")
x = tmp[:,0] + 1j * tmp[:,1]
y = tmp[:,2] + 1j * tmp[:,3]
line = tmp[:,4].astype(int)
fig, (ax1, ax2) = plt.subplots(ncols=2)
ax1.set_aspect(1)
ax2.set_aspect(1)

for i in tqdm(np.unique(line)):
    m = line == i
    x_line = x[m]
    y_line = y[m]
    if i % 3 == 0:
        ax1.plot(x_line.real, x_line.imag, c='tab:blue', linewidth=1, zorder=1)
        ax2.plot(y_line.real, y_line.imag, c='tab:blue', linewidth=1, zorder=1)
    elif i % 3 == 1:
        vertex_line_color = "tab:blue"
        # vertex_line_color = "tab:orange"
        ax1.plot(x_line.real, x_line.imag, c=vertex_line_color, linewidth=1, zorder=2)
        ax2.plot(y_line.real, y_line.imag, c=vertex_line_color, linewidth=1, zorder = 2)
    else:
        ax1.plot(x_line.real, x_line.imag, c='tab:green', linewidth=1, zorder=0)
        ax2.plot(y_line.real, y_line.imag, c='tab:green', linewidth=1, zorder=0)

plt.tight_layout()
# plt.savefig("conformal_L.png", dpi=150)
# plt.savefig("conformal_example.png", dpi=150)
```
